```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value']    
```

```python
def currency_name(curr_id):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    name = [response[curr]['Name'] for curr in response.keys() if response[curr]['ID'] == curr_id]   
    return name[0]

currency_name('R01700J')
#=> 'Турецких лир' 
```

```python
def wiki_header(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    return page.find('h1').text

wiki_header('https://en.wikipedia.org/wiki/Operating_system')
```

```python
def get_actors(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    a = page.find('ul', class_='styles_list__I97eu')
    return a
```

```python
def get_smm_index(group_name, token):
    url = 'https://api.vk.com/method/groups.getMembers'  
    params = {  
        'group_id': 'vk',  
        'v': 5.95,  
        'access_token': token  
    }  
    response = requests.get(url, params = params)  
    n_people = response.json()['response']['count'] 
    
    url = 'https://api.vk.com/method/wall.get'  
    params = {  
        'domain': group_name,   
        'access_token': token,  
        'v': 5.95  
    }  
    response = requests.get(url, params = params)  
    sum_com_lik_rep = 0
    for record in response.json()['response']['items'][:10]:
        sum_com_lik_rep += sum([record['comments']['count'], record['likes']['count'], record['reposts']['count']])
    return sum_com_lik_rep / n_people

get_smm_index('vk', token)
# => 0.00337003
```
